```python
import datetime
from pathlib import Path

import pandas as pd
import torch
from tqdm import tqdm
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer


from rouge_score import rouge_scorer

from bert_score import score as bert_score




def kl_divergence(p, q):
    """Compute the KL divergence between two distributions."""
    return torch.nan_to_num(p * (p / q).log(), nan=0.0).sum(-1)

def jensen_shannon_divergence(p, q):
    """Compute the Jensen-Shannon divergence between two distributions."""
    m = 0.5 * (p + q)
    return 0.5 * (kl_divergence(p, m) + kl_divergence(q, m))

class RSAReranking:
    """
    Re-ranks a list of candidate summaries using RSA.

    This implementation follows your original code: it computes a likelihood matrix
    (the "initial speaker probabilities"), then defines S(0)=initial likelihoods and
    L(0)=log_softmax(initial likelihoods along axis 0), and iterates with a rationality
    parameter. Finally, it selects the candidate with the highest speaker score.
    """
    def __init__(self, model, tokenizer, candidates, source_texts, batch_size=32, rationality=1, device="cpu"):
        self.model = model
        self.device = device
        self.tokenizer = tokenizer
        self.candidates = candidates
        self.source_texts = source_texts
        self.batch_size = batch_size
        self.rationality = rationality

    def compute_conditionned_likelihood(self, x, y, mean=True):
        """
        Compute the likelihood of y given x.
        We explicitly limit the length for candidate summaries (y) to avoid index errors.
        """
        assert len(x) == len(y), "x and y must have the same length"
        loss_fn = torch.nn.CrossEntropyLoss(reduction="none")
        batch_size = len(x)
        x_enc = self.tokenizer(x, return_tensors="pt", padding=True, truncation=True, max_length=1024)

        y_enc = self.tokenizer(y, return_tensors="pt", padding=True, truncation=True, max_length=128)
        x_ids = x_enc.input_ids.to(self.device)
        y_ids = y_enc.input_ids.to(self.device)
        outputs = self.model(
            input_ids=x_ids,
            decoder_input_ids=y_ids,
            attention_mask=x_enc.attention_mask.to(self.device),
            decoder_attention_mask=y_enc.attention_mask.to(self.device),
        )
        logits = outputs.logits
        shifted_logits = logits[..., :-1, :].contiguous()
        shifted_ids = y_ids[..., 1:].contiguous()
        likelihood = -loss_fn(shifted_logits.view(-1, shifted_logits.size(-1)), shifted_ids.view(-1))
        likelihood = likelihood.view(batch_size, -1).sum(-1)
        if mean:
            non_pad = (y_ids != self.tokenizer.pad_token_id).float().sum(-1)
            likelihood /= non_pad
        return likelihood

    def score(self, x, y, **kwargs):
        return self.compute_conditionned_likelihood(x, y, **kwargs)

    def likelihood_matrix(self):

        likelihood_matrix = torch.zeros((len(self.source_texts), len(self.candidates))).to(self.device)
        pairs = []
        for i, source_text in enumerate(self.source_texts):
            for j, candidate in enumerate(self.candidates):
                pairs.append((i, j, source_text, candidate))
        batches = [pairs[i: i + self.batch_size] for i in range(0, len(pairs), self.batch_size)]
        for batch in tqdm(batches, desc="Computing RSA likelihood matrix", leave=False):
            source_batch = [pair[2] for pair in batch]
            candidate_batch = [pair[3] for pair in batch]
            with torch.no_grad():
                likelihoods = self.score(source_batch, candidate_batch, mean=True)
            for k, (i, j, _, _) in enumerate(batch):
                likelihood_matrix[i, j] = likelihoods[k].detach()
        return likelihood_matrix

    def S(self, t):
        if t == 0:
            return self.initial_speaker_probas
        else:
            listener = self.L(t - 1)
            prod = listener * self.rationality
            return torch.log_softmax(prod, dim=-1)

    def L(self, t):
        speaker = self.S(t)
        return torch.log_softmax(speaker, dim=-2)

    def mk_listener_dataframe(self, t):
        self.initial_speaker_probas = self.likelihood_matrix()
        initial_listener_probas = self.L(0)
        uniform_distribution = torch.ones_like(initial_listener_probas) / len(self.source_texts)
        initial_consensuality_score = (torch.exp(initial_listener_probas) *
                                       (initial_listener_probas - torch.log(uniform_distribution))).sum(0).cpu().numpy()
        initial_listener_probas_np = initial_listener_probas.cpu().numpy()
        initial_listener_probas_df = pd.DataFrame(
            initial_listener_probas_np, index=self.source_texts, columns=self.candidates
        )
        initial_speaker_probas_np = self.S(0).cpu().numpy()
        initial_speaker_probas_df = pd.DataFrame(
            initial_speaker_probas_np, index=self.source_texts, columns=self.candidates
        )
        listener_df = pd.DataFrame(
            self.L(t).cpu().numpy(), index=self.source_texts, columns=self.candidates
        )
        consensuality_scores = (torch.exp(self.L(t)) *
                                (self.L(t) - torch.log(uniform_distribution))).sum(0).cpu().numpy()
        consensuality_scores_series = pd.Series(consensuality_scores, index=self.candidates)
        speaker_df = pd.DataFrame(
            self.S(t).cpu().numpy(), index=self.source_texts, columns=self.candidates
        )
        return listener_df, speaker_df, initial_listener_probas_df, initial_speaker_probas_df, initial_consensuality_score, consensuality_scores_series

    def rerank(self, t=1):
        """Re-ranks the candidates using RSA and returns several values."""
        listener_df, speaker_df, initial_listener_df, initial_speaker_df, init_cons_score, cons_scores = self.mk_listener_dataframe(t=t)
        best_rsa = speaker_df.idxmax(axis=1).values
        best_base = initial_listener_df.idxmax(axis=1).values
        return best_rsa, best_base, speaker_df, listener_df, initial_listener_df, initial_speaker_df, init_cons_score, cons_scores



def generate_summary(model, tokenizer, text, device, max_new_tokens=50, num_beams=4):

    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=1024).to(device)
    if "pegasus" in model.config.model_type.lower():

        model.config.min_length = 0
        model.config.max_length = max_new_tokens
    with torch.no_grad():
        generated_ids = model.generate(
            **inputs,
            max_length=max_new_tokens,
            min_length=0,
            num_beams=num_beams,
            early_stopping=True,
            eos_token_id=tokenizer.eos_token_id
        )
    summary = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    return summary


def main():

    dataset_path = Path("/content/glimpse-mds/data/processed/all_reviews_2017.csv")
    output_dir = Path("output")
    output_dir.mkdir(parents=True, exist_ok=True)

    # Candidate summarization models
    models_list = [
        "facebook/bart-large-cnn",
        "sshleifer/distilbart-cnn-12-6",
        "Falconsai/text_summarization"
    ]

    score_model_name = "facebook/bart-large-cnn"


    max_new_tokens = 50
    num_beams = 4
    rationality = 3.0

    device = "cpu"


    df = pd.read_csv(dataset_path)
    df = df.iloc[:314]
    for col in ['id', 'text', 'gold']:
        if col not in df.columns:
            raise ValueError(f"Input CSV must contain '{col}' column.")


    gen_models = {}
    print("Loading candidate generation models...")
    for model_name in models_list:
        print(f"  Loading {model_name} ...")
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        # Ensure pad_token and eos_token are defined
        if tokenizer.pad_token is None:
            tokenizer.pad_token = tokenizer.eos_token if tokenizer.eos_token is not None else "</s>"
        if tokenizer.eos_token is None:
            tokenizer.eos_token = tokenizer.pad_token
        if tokenizer.eos_token_id is None:
            tokenizer.eos_token_id = tokenizer.pad_token_id
        model.to(device)
        gen_models[model_name] = (model, tokenizer)


    print(f"Loading RSA scoring model: {score_model_name} ...")
    score_model = AutoModelForSeq2SeqLM.from_pretrained(score_model_name)
    score_tokenizer = AutoTokenizer.from_pretrained(score_model_name)
    if score_tokenizer.pad_token is None:
        score_tokenizer.pad_token = score_tokenizer.eos_token if score_tokenizer.eos_token is not None else "</s>"
    if score_tokenizer.eos_token is None:
        score_tokenizer.eos_token = score_tokenizer.pad_token
    if score_tokenizer.eos_token_id is None:
        score_tokenizer.eos_token_id = score_tokenizer.pad_token_id
    score_model.to(device)


    rouge = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    results = []
    print("Processing texts...")
    for _, row in tqdm(df.iterrows(), total=len(df)):
        sample_id = row["id"]
        text = row["text"]
        gold = row["gold"]
        candidate_summaries = {}


        for model_name, (model, tokenizer) in gen_models.items():
            summary = generate_summary(model, tokenizer, text, device, max_new_tokens=max_new_tokens, num_beams=num_beams)
            candidate_summaries[model_name] = summary


        candidates = [candidate_summaries[m] for m in models_list]


        reranker = RSAReranking(
            score_model,
            score_tokenizer,
            candidates=candidates,
            source_texts=[text],
            batch_size=1,
            rationality=rationality,
            device=device,
        )
        best_rsa, best_base, speaker_df, listener_df, _, _, _, cons_scores = reranker.rerank(t=1)
        best_candidate = best_rsa[0]
        best_model_by_rsa = next((m for m, summ in candidate_summaries.items() if summ == best_candidate), None)

        rsa_scores = {m: float(speaker_df.iloc[0, i]) for i, m in enumerate(models_list)}

        uniqueness_scores = {m: float(cons_scores[i]) for i, m in enumerate(models_list)}
        best_model_by_uniqueness = max(uniqueness_scores, key=uniqueness_scores.get)


        rouge_scores = rouge.score(gold, best_candidate)
        rouge_flat = {f"rouge_{k}": v.fmeasure for k, v in rouge_scores.items()}


        P, R, F1 = bert_score([best_candidate], [gold], lang='en', verbose=False, device='cpu')
        bertscore_dict = {
            "bertscore_precision": float(P.mean()),
            "bertscore_recall": float(R.mean()),
            "bertscore_f1": float(F1.mean())
        }

        results.append({
            "id": sample_id,
            "text": text,
            "gold": gold,
            "best_summary": best_candidate,
            "best_model_by_rsa": best_model_by_rsa,
            "best_model_by_uniqueness": best_model_by_uniqueness,
            "rsa_scores": rsa_scores,             # RSA speaker scores per model
            "uniqueness_scores": uniqueness_scores,  # Uniqueness scores per model
            **{f"summary_{m}": s for m, s in candidate_summaries.items()},
            **rouge_flat,
            **bertscore_dict,
        })

    now = datetime.datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    output_file = output_dir / f"multi_model_rsa_{now}.csv"
    pd.DataFrame(results).to_csv(output_file, index=False, encoding="utf-8")
    print(f"Output saved to {output_file}")

if __name__ == "__main__":
    main()
```
