```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from bertopic import BERTopic
from hdbscan import HDBSCAN
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, PegasusTokenizer
from pathlib import Path
from tqdm import tqdm
import json

from rsasumm.beam_search import RSAContextualDecoding
from rsasumm.rsa_reranker import RSAReranking

def get_aspect_prompt(topic, topic_model, n=20):

    topic_info = topic_model.get_topic(topic)
    if not topic_info:
        return ""
    keywords = [word for word, score in topic_info[:n]]
    return ", ".join(keywords)

def generate_topics(df, nr_topics):

    nltk.download("stopwords")
    stop_words = stopwords.words("english")
    vectorizer_model = CountVectorizer(stop_words=stop_words)
    # Set HDBSCAN parameters tuned for your dataset.
    hdbscan_model = HDBSCAN(min_cluster_size=15, min_samples=5)

    topic_model = BERTopic(
        verbose=True,
        vectorizer_model=vectorizer_model,
        hdbscan_model=hdbscan_model,
        nr_topics=nr_topics
    )

    topics, _ = topic_model.fit_transform(df["text"].tolist())
    df["aspect"] = topics
    return topic_model

def generate_summaries(df, model, tokenizer, device, decoding_config, topic_model, top_n_words=20):

    rsa = RSAContextualDecoding(model, tokenizer, device=device)
    summary_records = []

    print("Generating summaries for all samples...")
    for idx, row in tqdm(df.iterrows(), total=len(df)):
        text = row["text"]
        aspect = row["aspect"]
        aspect_prompt = get_aspect_prompt(aspect, topic_model, n=top_n_words)
        combined_text = f"Aspect: {aspect_prompt}\n{text}"

        inputs = tokenizer(
            combined_text,
            max_length=1024,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )
        inputs = {k: v.to(device) for k, v in inputs.items()}

        output = rsa.generate(
            target_id=0,
            source_texts_ids=inputs["input_ids"],
            source_text_attention_mask=inputs["attention_mask"],
            max_length=decoding_config.get("max_new_tokens", 50),
            top_p=decoding_config.get("top_p", 0.95),
            do_sample=decoding_config.get("do_sample", True),
            rationality=decoding_config.get("rationality", 8.0),
            temperature=decoding_config.get("temperature", 1.0),
            process_logits_before_rsa=decoding_config.get("process_logits_before_rsa", True)
        )
        summary = tokenizer.decode(output[0][0], skip_special_tokens=True)

        summary_records.append({
            "index": idx,
            "id": row["id"],
            "text": text,
            "gold": row["gold"],
            "summary": summary,
            "id_candidate": 0
        })
    return pd.DataFrame(summary_records)

def truncate_text(text, tokenizer, max_length):

    tokens = tokenizer.encode(text, max_length=max_length, truncation=True)
    return tokenizer.decode(tokens, skip_special_tokens=True)

def rerank_summaries(summaries_csv, model, tokenizer, device, rsa_iterations, rationality):

    df = pd.read_csv(summaries_csv)
    results = []
    for name, group in tqdm(df.groupby("id")):
        rsa_reranker = RSAReranking(
            model,
            tokenizer,
            candidates=group.summary.unique().tolist(),
            source_texts=group.text.unique().tolist(),
            batch_size=32,
            rationality=rationality,
            device=device
        )
        (best_rsa, best_base, speaker_df, listener_df, initial_listener,
         language_model_proba_df, initial_consensuality_scores, consensuality_scores) = rsa_reranker.rerank(t=2)
        gold = group["gold"].iloc[0]
        results.append({
            "id": name,
            "best_rsa": best_rsa,
            "best_base": best_base,
            "initial_consensuality_scores": initial_consensuality_scores,
            "consensuality_scores": consensuality_scores,
            "language_model_proba": language_model_proba_df.to_json() if hasattr(language_model_proba_df, "to_json") else str(language_model_proba_df),
            "gold": gold
        })
    return results

def main():

    dataset_path = Path("/content/glimpse-mds/data/processed/all_reviews_2017.csv")
    output_dir = Path("output")
    model_name = "facebook/bart-large-cnn"
    device = "cuda" if torch.cuda.is_available() else "cpu"
    max_new_tokens = 50
    nr_topics = 10
    top_n_words = 20
    do_sample = True
    top_p = 0.95
    temperature = 1.0
    rationality = 3.0
    rsa_iterations = 3
    rerank_flag = True


    output_dir.mkdir(parents=True, exist_ok=True)

    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    if "pegasus" in model_name.lower():
        tokenizer = PegasusTokenizer.from_pretrained(model_name)
    else:
        tokenizer = AutoTokenizer.from_pretrained(model_name)

    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    model = model.to(device)


    df = pd.read_csv(dataset_path)

    df = df.iloc[:314]

    topic_model = generate_topics(df, nr_topics=nr_topics)


    decoding_config = {
        "max_new_tokens": max_new_tokens,
        "do_sample": do_sample,
        "top_p": top_p,
        "temperature": temperature,
        "rationality": rationality,
        "process_logits_before_rsa": True
    }
    gen_df = generate_summaries(df, model, tokenizer, device, decoding_config, topic_model, top_n_words=top_n_words)
    summaries_csv = output_dir / f"{dataset_path.stem}_generated.csv"
    gen_df.to_csv(summaries_csv, index=False)
    print(f"Generated summaries saved to {summaries_csv}")


    gen_df = pd.read_csv(summaries_csv)

    gen_df["summary"] = gen_df["summary"].apply(lambda x: truncate_text(x, tokenizer, max_length=50))
    gen_df["text"] = gen_df["text"].apply(lambda x: truncate_text(x, tokenizer, max_length=200))
    gen_df.to_csv(summaries_csv, index=False)


    if rerank_flag:
        results = rerank_summaries(summaries_csv, model, tokenizer, device, rsa_iterations=rsa_iterations, rationality=3)
        rerank_records = []
        for res in results:
            rerank_records.append({
                "id": res["id"],
                "best_rsa": res["best_rsa"],
                "best_base": res["best_base"],
                "initial_consensuality_scores": json.dumps(res["initial_consensuality_scores"]) if isinstance(res["initial_consensuality_scores"], (list, dict)) else res["initial_consensuality_scores"],
                "consensuality_scores": json.dumps(res["consensuality_scores"]) if isinstance(res["consensuality_scores"], (list, dict)) else res["consensuality_scores"],
                "language_model_proba": res["language_model_proba"],
                "gold": res["gold"]
            })
        rerank_df = pd.DataFrame(rerank_records)
        rerank_csv = output_dir / f"{dataset_path.stem}_rsa_reranked.csv"
        rerank_df.to_csv(rerank_csv, index=False)
        print(f"Re-ranking results saved to {rerank_csv}")

if __name__ == "__main__":
    main()
```
